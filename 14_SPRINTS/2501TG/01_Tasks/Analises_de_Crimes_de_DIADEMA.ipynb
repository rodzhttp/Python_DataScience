### Totais de registros criminais - DataFrame

```python
from google.colab import drive
import pandas as pd
import numpy as np

# Monta o Google Drive
drive.mount('/content/drive')

# Caminho para o arquivo no Google Drive
caminho_do_arquivo = '/content/drive/My Drive/Colab Notebooks/SENAI - Python Para Data Science/ocorrencia_criminal_2022-2025.xlsx'

# Ler todas as planilhas do arquivo para um dicionário
dicionario_de_planilhas = pd.read_excel(caminho_do_arquivo, sheet_name=None)

# Criar uma lista vazia para guardar os DataFrames modificados
lista_de_dataframes = []

# Passar por cada planilha, adicionar a coluna 'Ano' e guardar na lista
print("Adicionando a coluna 'Ano' em cada planilha...")
for nome_da_planilha, df_da_planilha in dicionario_de_planilhas.items():
    # 'nome_da_planilha' será '2022', '2023', etc.
    df_da_planilha['Ano'] = nome_da_planilha

    # Adicionamos o DataFrame modificado à nossa lista
    lista_de_dataframes.append(df_da_planilha)

# Concatenar todos os DataFrames da lista em um único DataFrame
df_completo = pd.concat(lista_de_dataframes, ignore_index=True)

# Função para converter valores para inteiros quando possível
def to_int(val):
    try:
        return int(val)
    except (ValueError, TypeError):
        return val

# Aplicar a conversão para todos os valores do DataFrame
df_inteiros = df_completo.applymap(to_int)

# Configurar o pandas para exibir todas as linhas e colunas (opcional)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# ===========================================
# SEPARAÇÃO POR ANO EM UM DICIONÁRIO
# ===========================================
# Cria um dicionário com os DataFrames por ano
dfs_por_ano = {ano: df_inteiros[df_inteiros["Ano"] == ano] for ano in df_inteiros["Ano"].unique()}

# Exemplo: como acessar os DataFrames de cada ano
print("Exibindo os DataFrames separados por ano:")
for ano, df_ano in dfs_por_ano.items():
    print(f"\nAno: {ano}")
    display(df_ano)
```

```python
# ===========================================
# RENOMEAR COLUNA 'Total' PARA 'Total (Antigo)'
# ===========================================

if "Total" in df_inteiros.columns:
    df_inteiros.rename(columns={"Total": "Total (Antigo)"}, inplace=True)
    print("✅ Coluna 'Total' renomeada para 'Total (Antigo)'.")

# ===========================================
# RECALCULAR UM TOTAL GERAL DE TODAS AS COLUNAS NUMÉRICAS
# ===========================================

# Listar colunas que devem ser excluídas da soma
colunas_excluidas = [col for col in df_inteiros.columns if "TOTAL" in col.upper()] + ["Ano", "Natureza"]

# Selecionar apenas colunas numéricas para somar
colunas_para_somar = [col for col in df_inteiros.columns if col not in colunas_excluidas]

print("\nColunas que serão somadas no TOTAL GERAL RECALCULADO:")
print(colunas_para_somar)

# Garantir que os dados usados na soma sejam numéricos
df_inteiros["TOTAL GERAL RECALCULADO"] = df_inteiros[colunas_para_somar].apply(pd.to_numeric, errors='coerce').fillna(0).sum(axis=1)

print("\n✅ TOTAL GERAL RECALCULADO criado com sucesso!")

# ===========================================
# CRIAR COLUNA DIFERENCA TOTAL
# ===========================================

df_inteiros["DIFERENCA TOTAL"] = (
    pd.to_numeric(df_inteiros["TOTAL GERAL RECALCULADO"], errors='coerce').fillna(0) -
    pd.to_numeric(df_inteiros["Total (Antigo)"], errors='coerce').fillna(0)
)

# Exibir os dados separados por Ano
for ano, grupo in df_inteiros.groupby('Ano'):
    print(f"\n===== Ano: {ano} =====")
    display(grupo[["Natureza", "Total (Antigo)", "TOTAL GERAL RECALCULADO", "DIFERENCA TOTAL"]])

```

## **Comparando Homicídio Doloso e Homicídio Culposo por Ano**

```python
# Listas das naturezas que deseja somar
naturezas_doloso = [
    'HOMICÍDIO DOLOSO',
    'HOMICÍDIO DOLOSO POR ACIDENTE DE TRÂNSITO'
]

naturezas_culposo = [
    'HOMICÍDIO CULPOSO OUTROS',
    'HOMICÍDIO CULPOSO POR ACIDENTE DE TRÂNSITO'
]

# Filtrar linhas para cada grupo
df_doloso = df_inteiros[df_inteiros['Natureza'].isin(naturezas_doloso)]
df_culposo = df_inteiros[df_inteiros['Natureza'].isin(naturezas_culposo)]

# Agrupar e somar por ano
total_doloso = df_doloso.groupby('Ano')['Total (Antigo)'].sum()
total_culposo = df_culposo.groupby('Ano')['Total (Antigo)'].sum()

# Criar DataFrame para plotagem
anos = sorted(set(total_doloso.index).union(total_culposo.index))
df_plot = pd.DataFrame({'Ano': anos})
df_plot['Total Homicídio Doloso'] = df_plot['Ano'].map(total_doloso).fillna(0).astype(int)
df_plot['Total Homicídio Culposo'] = df_plot['Ano'].map(total_culposo).fillna(0).astype(int)

# Calcular redução percentual em relação a 2022
base_doloso = df_plot.loc[df_plot['Ano'] == 2022, 'Total Homicídio Doloso'].values[0]
base_culposo = df_plot.loc[df_plot['Ano'] == 2022, 'Total Homicídio Culposo'].values[0]
df_plot['Redução Doloso (%)'] = ((base_doloso - df_plot['Total Homicídio Doloso']) / base_doloso * 100).round(1)
df_plot['Redução Culposo (%)'] = ((base_culposo - df_plot['Total Homicídio Culposo']) / base_culposo * 100).round(1)

# Plotar gráfico
plt.style.use('ggplot')
plt.figure(figsize=(10,6))

width = 0.35
x = df_plot['Ano']

bars1 = plt.bar(x - width/2, df_plot['Total Homicídio Doloso'], width, label='Homicídio Doloso', color='#d62728', alpha=0.8)
bars2 = plt.bar(x + width/2, df_plot['Total Homicídio Culposo'], width, label='Homicídio Culposo', color='#1f77b4', alpha=0.8)

# Valores nas barras
for i in range(len(df_plot)):
    plt.text(x[i] - width/2, df_plot['Total Homicídio Doloso'][i] + 0.3,
             df_plot['Total Homicídio Doloso'][i], ha='center', fontsize=9)
    plt.text(x[i] + width/2, df_plot['Total Homicídio Culposo'][i] + 0.3,
             df_plot['Total Homicídio Culposo'][i], ha='center', fontsize=9)

# Linhas de tendência pontilhadas
plt.plot(x, df_plot['Total Homicídio Doloso'], color='#d62728', linestyle='--', alpha=0.6)
plt.plot(x, df_plot['Total Homicídio Culposo'], color='#1f77b4', linestyle='--', alpha=0.6)

plt.title('Comparação entre Homicídios Dolosos e Culposos por Ano', pad=15)
plt.xlabel('Ano')
plt.ylabel('Número de Ocorrências')
plt.xticks(x)
plt.legend()

plt.tight_layout()
plt.show()

# Mostrar tabela comparativa com redução percentual
print("\nTabela Comparativa Anual:")
print(df_plot.to_string(index=False))
```

## **Comparando Roubos e Furtos de Veiculos**

```python
# Listas das naturezas que queremos somar
naturezas_roubo = ['TOTAL DE ROUBO - OUTROS (1)', 'ROUBO DE VEÍCULO']
naturezas_furto = ['FURTO - OUTROS', 'FURTO DE VEÍCULO']

# Filtrar linhas correspondentes a cada grupo
df_roubo = df_inteiros[df_inteiros['Natureza'].isin(naturezas_roubo)]
df_furto = df_inteiros[df_inteiros['Natureza'].isin(naturezas_furto)]

# Agrupar por ano e somar os totais recalculados
total_roubo = df_roubo.groupby('Ano')['TOTAL GERAL RECALCULADO'].sum()
total_furto = df_furto.groupby('Ano')['TOTAL GERAL RECALCULADO'].sum()

# Criar DataFrame para plotagem
anos = sorted(set(total_roubo.index).union(total_furto.index))
df_plot = pd.DataFrame({'Ano': anos})
df_plot['Total Roubo'] = df_plot['Ano'].map(total_roubo).fillna(0).astype(int)
df_plot['Total Furto'] = df_plot['Ano'].map(total_furto).fillna(0).astype(int)

# Calcular redução percentual em relação a 2022
base_roubo = df_plot.loc[df_plot['Ano'] == 2022, 'Total Roubo'].values[0]
base_furto = df_plot.loc[df_plot['Ano'] == 2022, 'Total Furto'].values[0]
df_plot['Redução Roubo (%)'] = ((base_roubo - df_plot['Total Roubo']) / base_roubo * 100).round(1)
df_plot['Redução Furto (%)'] = ((base_furto - df_plot['Total Furto']) / base_furto * 100).round(1)

# Plotando o gráfico
plt.style.use('ggplot')
plt.figure(figsize=(12,7))

width = 0.35
x = df_plot['Ano']

bars1 = plt.bar(x - width/2, df_plot['Total Roubo'], width,
                label='Roubo (Outros + Veículo)', color='#d62728', alpha=0.8)
bars2 = plt.bar(x + width/2, df_plot['Total Furto'], width,
                label='Furto (Outros + Veículo)', color='#1f77b4', alpha=0.8)

plt.title('Comparação entre Roubos e Furtos (2022-2025)', pad=20, fontsize=14)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Número de Ocorrências', fontsize=12)
plt.xticks(x, fontsize=11)
plt.yticks(fontsize=11)
plt.legend(fontsize=12)

# Valores nas barras
for bar in bars1:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}', ha='center', va='bottom', fontsize=10)

for bar in bars2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}', ha='center', va='bottom', fontsize=10)

# Linha de tendência
plt.plot(x, df_plot['Total Roubo'], color='#d62728', linestyle='--', alpha=0.5)
plt.plot(x, df_plot['Total Furto'], color='#1f77b4', linestyle='--', alpha=0.5)

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Mostrar tabela comparativa com redução percentual
print("\nTabela Comparativa Anual:")
print(df_plot.to_string(index=False))
```

## **Comparando Lesões Corporais Dolosas e Culposas**

```python
# Listas das naturezas que você quer comparar
naturezas_dolosa = [
    'LESÃO CORPORAL DOLOSA'
]

naturezas_culposa = [
    'LESÃO CORPORAL CULPOSA - OUTRAS',
    'LESÃO CORPORAL CULPOSA POR ACIDENTE DE TRÂNSITO'
]

# Filtrar linhas correspondentes a cada grupo
df_dolosa = df_inteiros[df_inteiros['Natureza'].isin(naturezas_dolosa)]
df_culposa = df_inteiros[df_inteiros['Natureza'].isin(naturezas_culposa)]

# Agrupar por ano e somar TOTAL GERAL RECALCULADO
total_dolosa = df_dolosa.groupby('Ano')['TOTAL GERAL RECALCULADO'].sum()
total_culposa = df_culposa.groupby('Ano')['TOTAL GERAL RECALCULADO'].sum()

# Criar DataFrame para plotagem
anos = sorted(set(total_dolosa.index).union(total_culposa.index))
df_plot = pd.DataFrame({'Ano': anos})
df_plot['Lesão Corporal Dolosa'] = df_plot['Ano'].map(total_dolosa).fillna(0).astype(int)
df_plot['Lesão Corporal Culposa'] = df_plot['Ano'].map(total_culposa).fillna(0).astype(int)

# Calcular redução percentual em relação a 2022
base_dolosa = df_plot.loc[df_plot['Ano'] == 2022, 'Lesão Corporal Dolosa'].values[0]
base_culposa = df_plot.loc[df_plot['Ano'] == 2022, 'Lesão Corporal Culposa'].values[0]
df_plot['Redução Dolosa (%)'] = ((base_dolosa - df_plot['Lesão Corporal Dolosa']) / base_dolosa * 100).round(1)
df_plot['Redução Culposa (%)'] = ((base_culposa - df_plot['Lesão Corporal Culposa']) / base_culposa * 100).round(1)

# Plotar o gráfico
plt.style.use('ggplot')
plt.figure(figsize=(12,7))

width = 0.35
x = df_plot['Ano']

bars1 = plt.bar(x - width/2, df_plot['Lesão Corporal Dolosa'], width,
                label='Lesão Corporal Dolosa', color='#d62728', alpha=0.8)
bars2 = plt.bar(x + width/2, df_plot['Lesão Corporal Culposa'], width,
                label='Lesão Corporal Culposa (Outras + Acidente de Trânsito)', color='#1f77b4', alpha=0.8)

# Valores nas barras
for bar in bars1:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}', ha='center', va='bottom', fontsize=10)

for bar in bars2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}', ha='center', va='bottom', fontsize=10)

# Linhas de tendência
plt.plot(x, df_plot['Lesão Corporal Dolosa'], color='#d62728', linestyle='--', alpha=0.6)
plt.plot(x, df_plot['Lesão Corporal Culposa'], color='#1f77b4', linestyle='--', alpha=0.6)

plt.title('Comparação entre Lesões Corporais Dolosas e Culposas por Ano', pad=20, fontsize=14)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Número de Ocorrências', fontsize=12)
plt.xticks(x, fontsize=11)
plt.yticks(fontsize=11)

# Legenda dentro do gráfico no canto inferior direito
plt.legend(fontsize=12, loc='lower right')

plt.tight_layout()
plt.show()

# Exibir tabela comparativa
print("\nTabela Comparativa Anual:")
print(df_plot.to_string(index=False))
```
